```python
#checking for null values in the df
train_df.apply(lambda x: sum(x.isnull()),axis=0)
```

Based on the output above, the Valencia_pressure column has 2068 null values. 
The most simple technique  to replace missing data is with some constant value. The value can be any number that seemed appropriate.

```python
#Calculate percentage of missing values in the Valencia_pressure column
train_df['Valencia_pressure'].isnull().mean() * 100
```

About 24% of the data in the column is missing, so we can procced to fill in the 
missing values ,as 24% still falls under acceptable perameters of %missing values in a df.
Theoretically, 25 to 30% is the maximum missing values that are allowed, beyond which we might want to drop the variable from analysis.

```python
train_df['Valencia_pressure'].median(), train_df['Valencia_pressure'].mean(), train_df['Valencia_pressure'].mode()
```

```python
# Distribution plot for deciding imputation technique
fig, ax = plt.subplots(figsize = (5, 4))
sns.kdeplot(train_df['Valencia_pressure'])
```

1. Replacing missing data by the mode is not common practice for numerical variables. 
2. If the variable is skewed, the mean is biased by the values at the far end of the distribution. 
Therefore, the median is a better representation of the majority of the values in the variable

```python
# Fill in missing values using the median value
train_df['Valencia_pressure'] = train_df['Valencia_pressure'].fillna(train_df['Valencia_pressure'].median())
```

```python
# Valencia_pressure column has no null values anymore
train_df['Valencia_pressure'].isnull().sum()
```

Having viewed the data types pf the dataframe, we saw that the time column had object as it's datatype.

```python
new_df = train_df.copy()
new_df['time']= pd.to_datetime(new_df['time'])
```

The time column has changed dtype to datetime, however, we note that two columns namely: Valencia_wind_deg & Seville_pressure still

```python
new_df[['Valencia_wind_deg', 'Seville_pressure']]
```

From the output we see that the columns contain string values, which, make no sense as they are supposed to be describing wind degrees and 
atmosphoric pressure respectively. So, we assume that it was a mistake done by the data collector. Therefore, we have the option to either remove the columns or convert them into numeric values that can be contributing variables to make our model. Below, we concert the column values to numbers. 

```python
new_df['Valencia_wind_deg'] = new_df['Valencia_wind_deg'].str.extract('(\d+)')
new_df['Seville_pressure'] = new_df['Seville_pressure'].str.extract('(\d+)')
new_df.head()
```

```python
new_df.info()
```

```python
# Convert the column values to floats
new_df['Valencia_wind_deg'] = new_df['Valencia_wind_deg'].astype(float)
new_df['Seville_pressure'] = new_df['Seville_pressure'].astype(float)
```
